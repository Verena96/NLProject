```python
#Empty dataframe to store pdf data
d = pd.DataFrame(columns = ['Name','Text','EconOutlook','Risks_Scenario','MonetaryPol','MonetPolDec'])

#Access all pdfs from path
files = [f for f in os.listdir("C:/Users/veren/Documentos/JPMorgan/Minutas_15_20") if os.path.isfile(f) and f.endswith('.pdf')]

for file in files:
    
    try:

        pdfFileObject = open(file, 'rb')
        pdfReader = PyPDF2.PdfFileReader(pdfFileObject)
        count = pdfReader.numPages
        x = ''

        for i in range(2,count):
            page = pdfReader.getPage(i)
            cont = page.extractText()
            x+=cont
        d = d.append({'Name':file.lower().split('.pdf')[0],'Text':x},ignore_index=True)  
        
    except:
        
        print('Arquivo com erro: ',file)

#Texts inserted in dataset
d = d.drop_duplicates()
d['Text']=d['Text'].apply(lambda x:split(x)).str.lower()
```

```python
d['EconOutlook'][3] = " ".join(d['Text'][3].split('.')[9:12])
d['Risks_Scenario'][3] = ".".join(d['Text'][3].split('.')[12:13]+d['Text'][3].split('.')[38:43])
d['MonetaryPol'][3] = ".".join(d['Text'][3].split('.')[43:46] +d['Text'][3].split('.')[46:51]) 
d['MonetPolDec'][3] = ".".join(d['Text'][3].split('.')[51:54])
```

```python
#Cleaning
d['Text'] = d['Text'].str.replace('-',' ')
d['Text'] = d['Text'].apply(lambda x:x.replace(x.split('the')[-1],''))
d['Text'] = d['Text'].apply(lambda x:x.replace('the following members of the',''))
```

```python
#Splitting the texts according to topic segmentation - this is a pattern from the texts released by the central bank
for i in range(0,len(d)):
    
    try:
        delimiters = "a)", "b)", "c)","d)"
        example = d['Text'][i]
        regexPattern = '|'.join(map(re.escape, delimiters))

        lista_t = re.split(regexPattern, example)
        d['EconOutlook'][i] = lista_t[1]
        d['Risks_Scenario'][i] = lista_t[2]
        d['MonetaryPol'][i] = lista_t[3]
        d['MonetPolDec'][i] = lista_t[4]
        
    except:
        
        pass
        
#Copy data for further changes
dc = d.copy()
```

```python
#Clean ponctuation
exclude = set(string.punctuation)
for col in ['Text', 'EconOutlook', 'Risks_Scenario', 'MonetaryPol','MonetPolDec']:
    
    dc[col] = dc[col].apply(lambda texto:''.join(ch for ch in texto if ch not in exclude))
```

```python
#Replacements from some common phrases in the texts
dc['MonetPolDec'] = dc['MonetPolDec'].apply(lambda x:x.split('decided')[-1])
dc['EconOutlook'] = dc['EconOutlook'].str.replace('update of economic outlook','')
dc['EconOutlook'] = dc['EconOutlook'].str.replace('baseline scenario','')
dc['Risks_Scenario'] = dc['Risks_Scenario'].str.replace('risks around the baseline inflation scenario','')
dc['Risks_Scenario'] = dc['Risks_Scenario'].str.replace('baseline scenario','')
dc['MonetaryPol'] = dc['MonetaryPol'].str.replace('discussion about the conduct of monetary policy','')
dc['MonetaryPol'] = dc['MonetaryPol'].str.replace('the committee members discussed the evolution of economic activity in light of available data.','')
```

```python
#More cleaning
for col in ['Text', 'EconOutlook', 'Risks_Scenario', 'MonetaryPol','MonetPolDec']:
    
    dc[col] = dc[col].apply(lambda texto: clean_ents(texto))
    dc[col] = dc[col].apply(lambda texto: clean_words(texto))
    dc[col] = dc[col].apply(lambda texto: sub_str(texto))  
```
